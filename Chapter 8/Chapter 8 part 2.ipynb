# Chapter 8 Exercise 9##
Solution to Programming Exercise 16 of Chapter 8 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
def ex9():
    odolist = []
    gaslist = []
    totalmiles = 0
    totalgas = 0
    sodoread = int(input("Enter starting odometer reading: "))
    codoread = input("Enter current odometer reading and gas consumed, separated by space: ")
    
    while codoread != "":
        codoreadstr = codoread.split(' ')
        odolist.append(codoreadstr[0])
        gaslist.append(codoreadstr[1])
        codoread = input("Enter current odometer reading and gas consumed, separated by space: ")
        
    for j in range(len(odolist) - 1):
        if j == 1:
            dist = int(odolist[j]) - sodoread
        else:
            dist = int(odolist[j + 1]) - int(odolist[j])
        totalmiles = totalmiles + dist
        totalgas = int(totalgas) + int(gaslist[j])
        print(str(j) + ' ' + str(dist / int(gaslist[j])))
    totalmpg = totalmiles / totalgas
    print("{0:3}".format(totalmpg))
        
ex9()    
```

```python
def ex11():
    heatingdegree = 0
    coolingdegree = 0
    avgtemp = int(input("What is the average temperature for a day? "))
    while avgtemp != "":
        if int(avgtemp) < 60:
            heatingdegree = 60 - int(avgtemp) + (heatingdegree)
        elif int(avgtemp) > 80:
            coolingdegree = int(avgtemp) - 80 + int(coolingdegree)
        else:
            print("It is neither heating degree day not cooling degree day")
        avgtemp = input("What is the average temperature for a day? ")
        
    print("The amount of heating degree days is:", heatingdegree, "The amount of cooling degree days is:", coolingdegree)
    
ex11()
```

# Chapter 8 Exercise 13##
Solution to Programming Exercise 16 of Chapter 8 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
from graphics import * 

def ex13():
    donerecx = 100
    donerecy = 450
    n = 0
    sx = 0
    sy = 0
    sxsq = 0
    sxy = 0
    win = GraphWin("Linear Regression Line", 500, 500)
    donerec = Rectangle(Point(50,450), Point(100,490))
    donerec.draw(win)
    Text(donerec.getCenter(), "Done").draw(win)
    
    p = win.getMouse()
    p.draw(win)
    x = p.getX()
    y = p.getY() 
    
    while (x >= donerecx or y <= donerecy):
        n = n + 1
        sx = sx + x
        sy = sy + y
        sxsq = sxsq + x**2
        sxy = sxy + x*y
        p = win.getMouse()
        p.draw(win)
        x = p.getX()
        y = p.getY()
        
    ybar = sy/n
    xbar = sx/n
    m = (sxy - n*xbar*ybar)/(sxsq - n*(xbar**2))
    x1 = 0
    y1 = ybar + m*(x1 - xbar)
    p1 = Point(x1,y1)
    x2 = 500
    y2 = ybar + m*(x2 - xbar)
    p2 = Point(x2,y2)
    line = Line(p1,p2)
    line.draw(win)

ex13()
```

# Chapter 8 Exercise 14##
Solution to Programming Exercise 16 of Chapter 8 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
from graphics import *


def origImage(picture):
    img = Image(Point(200,250), picture)
    width = img.getWidth()
    height = img.getHeight()
    win = GraphWin(picture, width, height)
    img.draw(win)
    return img

def Grey(img):
    for rownum in range(img.getHeight()):
        for columnnum in range(img.getWidth()):
            r, g, b = img.getPixel(columnnum, rownum)
            brightness = int(round(0.299 * r + 0.587 * g + 0.114 * b))
            img.setPixel(columnnum, rownum, color_rgb(brightness, brightness, brightness))
        update()
            

def ex14():
    os.chdir('C:\\Users\\autej\\Documents\\Zelle_Python')
    
    File = input("Enter the name of the file: ")
    image = origImage(File)
    
    Grey(image)
    
    newpic = input("What do you want to name your new picture: ")
    image.save(newpic)

ex14()
```

## Chapter 8 Exercise 16

Solution to Programming Exercise 16 of Chapter 8 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
from graphics import *

def handleKey(k, win):
    if k == "r":
        win.setBackground("pink")
    elif k == "w":
        win.setBackground("white")
    elif k == "g":
        win.setBackground("lightgray")
    elif k == "b":
        win.setBackground("lightblue")
        
def handleClick(pt, win):
    entry = Entry(pt, 10)
    entry.draw(win)
    
    while True:
        key = win.getKey()
        if key == "Return":
            Text(pt, entry.getText()).draw(win)
            break
        elif key == "Escape":
            break
        else:
            pass
        
    entry.undraw()
    win.checkMouse()
    
    
    
def ex16():
    
    win = GraphWin("Click and Type", 500, 500)
    
    while True:
        key = win.checkKey()
        if key == "q":
            break
            
        if key:
            handleKey(key, win)
            
        pt = win.checkMouse()
        if pt:
            handleClick(pt, win)
            
    win.close()
    
ex16()
```
