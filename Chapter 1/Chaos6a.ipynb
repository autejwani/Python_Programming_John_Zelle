```
# File: chaos6a.py
# A simple program illustrating chaotic behavior.  It prints out 100 values.
# Solution to Programming Exercise 6a of Chapter 1 of Python Programming:
# An Introduction to Computer Science, 3rd Edition, Franklin, Beedle, and 
# Associates

def main():
    print("This program illustrates a chaotic function")
    x = eval(input("Enter a number between 0 and 1: "))
    for i in range(100):
        x = 3.9 * x * (1 - x)
        print(x)

main()
```
