# Chapter 7 Exercise 14##
Solution to Programming Exercise 14 of Chapter 7 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
from graphics import *
import math

def ex14():

    radius = float(input("Please enter your radius: "))
    yintercept = float(input("Please enter your y-intercept: "))

    win = GraphWin("Circle Intersection")
    win.setCoords(-10,-10,10,10)

    Circle(Point(0,0), radius).draw(win)
    Line(Point(-10, yintercept), Point(10, yintercept)).draw(win)

    if abs(yintercept) <= radius:
        x = math.sqrt(radius * radius - yintercept * yintercept)
        print("X values of intersection", -x, x)

        p1 = Circle(Point(x,yintercept), 0.25)
        p1.setOutline("red")
        p1.setFill("red")
        p1.draw(win)

        p2 = p1.clone()
        p2.move(-2 * x, 0)
        p2.draw(win)
    else:
        print("The line is not intersecting the circle.")

    print("Click the window to quit")
    win.getMouse()
    win.close()

if __name__ == '__main__':
    ex14()

```

# Chapter 7 Exercise 15##
Solution to Programming Exercise 15 of Chapter 7 of Python Programming: An Introduction to Computer Science, John Zelle, 3rd Edition, Franklin, Beedle, and Associates

```python
import math

from graphics import *

def ex15():
    win = GraphWin("Line Segment Info", 400, 400)
    win.setCoords(-10,-10,10,10)

    msg = Text(Point(0,-9.5), "Click on endpoints of a line segment.")
    msg.draw(win)

    p1 = win.getMouse()
    p1.draw(win)

    p2 = win.getMouse()
    p2.draw(win)

    line = Line(p1,p2)
    line.draw(win)

    mark = Circle(line.getCenter(),0.15)
    mark.setFill("pink")
    mark.draw(win)

    dx = p2.getX() - p1.getX()
    dy = p2.getY() - p1.getY()
    if dx != 0.0:
        slope = str(round(dy/dx,2))
    else:
        slope = "inf"
    length = str(round(math.sqrt(dx*dx + dy*dy),2))

    msg.setText("Length: {}, Slope: {}".format(length, slope))
    win.getMouse()
    win.close()


if __name__ == '__main__':
    ex15()
```
