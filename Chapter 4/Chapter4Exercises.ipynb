**Exercise 6**

```python
from graphics import*

win = GraphWin("Investment Growth Chart", 500, 500)
win.setBackground("white")
    
def ex6():
    # Introduction
    print("This program plots the growth of a ten year investment. ")
    
    # Draw interface and get principal and intrest rate
    
    Text(Point(120,280), "Enter the initial principal ").draw(win)
    inputText1 = Entry(Point(260,280), 10)
    inputText1.setText("0.0")
    inputText1.draw(win)
    Text(Point(120,315), "Enter the annualized interest rate ").draw(win)
    inputText2 = Entry(Point(290,315), 10)
    inputText2.setText("0.0")
    inputText2.draw(win)
    win.getMouse()
    principal = float(inputText1.getText())
    apr = float(inputText2.getText())
    
    # Create a graphics window with labels on left edge
    
    Text(Point(20,230), ' 0.0K').draw(win)
    Text(Point(20,180), ' 2.5K').draw(win)
    Text(Point(20,130), ' 5.0K').draw(win)
    Text(Point(20,80), ' 7.5K').draw(win)
    Text(Point(20,30), ' 10.0K').draw(win)
    
    # Draw bar for initial principal
    height = principal * 0.02
    bar = Rectangle(Point(40,230), Point(65,230-height))
    bar.setFill("green")
    bar.setWidth(2)
    bar.draw(win)
    
    # Draw bars for successive years
    for year in range(1,11):
        # calculate value for the next year
        principal = principal * (1 + apr)
        # draw bar for this value
        x11 = year * 25 + 40
        height = principal * 0.02
        bar = Rectangle(Point(x11, 230), Point(x11+25, 230-height))
        bar.setFill("green")
        bar.setWidth(2)
        bar.draw(win)
        
        
    input("Press <Enter> to quit")
    win.close()
    
ex6()
```

```python
from graphics import *
import math

win = GraphWin("Exercise 7 Program", 500, 500)
win.setBackground("white")

def drawCircle(x,y,r):
    circle = Circle(Point(x,y),r)
    #circle.setFill("black")
    circle.draw(win)

def drawRedDot(x,y):
    circle = Circle(Point(x,y),5)
    circle.setFill("red")
    circle.draw(win)
    
def drawInt(y):
    line = Line(Point(200,y),Point(300,y)).draw(win)
    
def ex7():
     # Draw interface and get principal and intrest rate
    
    Text(Point(75,50), "Enter the radius ").draw(win)
    inputText1 = Entry(Point(200,50), 10)
    inputText1.setText("0.0")
    inputText1.draw(win)
    Text(Point(80,100), "Enter the intercept ").draw(win)
    inputText2 = Entry(Point(200,100), 10)
    inputText2.setText("0.0")
    inputText2.draw(win)
    win.getMouse()
    radius = float(inputText1.getText())
    intercept = float(inputText2.getText())
    
    ## Origin is mapped to point(250,250)
    xaxis = Line(Point(200,250), Point(300,250)).draw(win)
    yaxis = Line(Point(250,300), Point(250,200)).draw(win)
    drawCircle(250,250,radius*5)
    y = 250 - intercept*5
    drawInt(y)
    if (abs(intercept) > radius):
        Text(Point(250,400),"Intercept is greater than radius.  There is no point of intersection.").draw(win)
    else:
        x = math.sqrt(((radius)**2)-(intercept**2))
        drawRedDot(250-5*x,y)
        drawRedDot(250+5*x,y)
        Text(Point(250,400),"x values of the point of intersection are +/-" + str(x)).draw(win)
    
    
    
    
    
ex7()
```

```python
from graphics import *
import math 

win = GraphWin("Line Segment Info", 500, 500)

def ex8():
    msg = Text(Point(230,100), "Click on endpoints of a line segment.")
    msg.draw(win)
    
    p1 = win.getMouse()
    p1.draw(win)
    
    p2 = win.getMouse()
    p2.draw(win)
    
    line = Line(p1,p2)
    line.draw(win)
    
    mp = Circle(line.getCenter(),3)
    mp.setFill("cyan")
    mp.draw(win)
    
    dx = p2.getX() - p1.getX()
    dy = p2.getY() - p1.getY()
    slope = float(dy)/dx
    length = math.sqrt(dx**2 + dy**2)
    Text(Point(250,400), "Length: " + "{:.2f}".format(length)).draw(win)
    Text(Point(250,420), "Slope: " + "{:.2f}".format(slope)).draw(win)

ex8()
```

**Exercise 10**

```python
from graphics import *
import math 

win = GraphWin("Triangle Information", 500, 500)

def calculateLength(p1,p2):
    x1 = p1.getX()
    y1 = p1.getY()
    x2 = p2.getX()
    y2 = p2.getY()
    length = math.sqrt((x1-x2)**2 + (y1 - y2)**2)
    return length
    
def ex10():
    msg = Text(Point(230,100), "Click for the vertices of a triangle.")
    msg.draw(win)
    
    pa = win.getMouse()
    pa.draw(win)
    
    pb = win.getMouse()
    pb.draw(win)
    
    pc = win.getMouse()
    pc.draw(win)
    
    Polygon(pa,pb,pc).draw(win)
    
    a = calculateLength(pb,pc)
    b = calculateLength(pa,pc)
    c = calculateLength(pa,pb)
    s = (a + b + c)/2
    
    area = math.sqrt(s*(s - a)*(s - b)*(s - c))
    perimeter = a + b + c
    
    Text(Point(300,400), "The perimeter is " + "{:.2f}".format(perimeter)).draw(win)
    Text(Point(300,450),"The area is " + "{:.2f}".format(area)).draw(win)
    
ex10()
```

**Excercise 11**

```python
from graphics import *

def main():
    win = GraphWin("Design a House",500,500)
    win.setCoords(0, 0, 200, 200)
    message = Text(Point(100,5),"")
    message.draw(win)

    # Draw the frame of the house
    message.setText("Click on low left corner of the Graph Window")
    frameLowLef = win.getMouse()
    frameLowLef.draw(win)
    message.setText("Click up right corner of the frame")
    frameUpRig = win.getMouse()
    frameLowLef.undraw()
    Rectangle(frameLowLef, frameUpRig).draw(win)

    # Draw a door
    houseWid = frameUpRig.getX() - frameLowLef.getX()
    doorWid = 0.2 * houseWid
    halfDoor = doorWid/2.0
    message.setText("Click on the center of the top of the door")
    door2 = win.getMouse()
    door2.move(halfDoor, 0)
    doorX1 = door2.getX() - doorWid
    doorY1 = frameLowLef.getY()
    door = Rectangle(Point(doorX1,doorY1), door2)
    door.setFill("red")
    door.draw(win)

    # Draw a window
    message.setText( "Click on the center of the window")
    windowCenter = win.getMouse()
    w1 = windowCenter.clone()
    winOff = 0.5 * halfDoor
    w1.move(-winOff,winOff)
    w2 = windowCenter.clone()
    w2.move(winOff, -winOff)
    Rectangle(w1,w2).draw(win)
    
    message.setText( "Click on the peak of the roof")
    point = win.getMouse()
    frameUpLef = frameUpRig.clone()
    frameUpLef.move(-houseWid,0)
    roof = Polygon(frameUpLef, point, frameUpRig)
    roof.draw(win)
    roof.setFill("black")
    
    message.setText( "Click anywhere to quit." )
    win.getMouse()
    win.close()
    
main()

```
